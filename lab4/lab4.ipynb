```python
class MultiLayerPerceptron:
    def __init__(self, input_size, hidden_size, output_size, w1=None, w2=None, bias=True):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.bias = bias
        if self.bias:
            self.input_size += 1
        # print(f'input_size: {self.input_size}, hidden_size: {self.hidden_size}, output_size: {self.output_size}')
        # if W1 is None initialize weights for hidden layer
        if w1 is None:
            self.W1 = np.random.randn(self.input_size, self.hidden_size)
        else:
            if self.bias:
                self.W1 = np.insert(w1, 0, 1, axis=0)
            else:
                self.W1 = w1
        # if W2 is None initialize weights for output layer
        if w2 is None:
            self.W2 = np.random.randn(self.hidden_size, self.output_size)
        else:
            self.W2 = w2
        
        # print(pd.DataFrame(self.W1))
        # print(pd.DataFrame(self.W2)) 
            
        self.learning_rate = 0.01
        
        # self.W1 = np.random.randn(self.input_size, self.hidden_size)
        # self.W2 = np.random.randn(self.hidden_size, self.output_size)
        # self.b1 = np.random.randn(self.hidden_size)
        # self.b2 = np.random.randn(self.output_size)
        # self.learning_rate = 0.01
        # self.loss = []
        # self.accuracy = []
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def fit(self, X, y):
        # initialize weights
        for _ in range(1):
            # if self.bias then prepend 1 to the end of X
            if self.bias:
                X = np.insert(X, 0, 1, axis=1)
            for idx, x_i in enumerate(X):
                # forward propagation
                # calculate output for hidden layer
                self.z1 = np.dot(x_i, self.W1)
                print(self.z1)
                self.a1 = self.sigmoid(self.z1)
                print(self.a1)
                # calculate output for output layer
                self.z2 = np.dot(self.a1, self.W2)
                print(self.z2)
                self.a2 = self.sigmoid(self.z2)
                print(self.a2)
                
                # back propagation
                # calculate error term, delta2, for output layer
                self.delta2 = self.a2 * (1 - self.a2) * (y[idx] - self.a2)
                print(f'delta2: {self.delta2}')
                # calculate error term, delta1, for hidden layer
                self.delta1 = self.a1 * (1 - self.a1) * np.dot(self.delta2, self.W2.T)
                print(f'delta1: {self.delta1}')
                # update weights
                self.W2 += self.learning_rate * self.delta2 * self.a2
                print(f'W2: {self.W2}')
                self.W1 += self.learning_rate * self.delta1 * self.a1 
                print(f'W1: {self.W1}')
        
    def predict(self, X):
        # pred = self.sigmoid(np.dot(self.sigmoid(np.dot(X, self.W1)), self.W2))
        # return pred
        if self.bias:
            X = np.insert(X, 0, 1, axis=1)
        pred = []
        for idx, x_i in enumerate(X):
            # forward propagation
            # calculate output for hidden layer
            self.z1 = np.dot(x_i, self.W1)
            print(self.z1)
            self.a1 = self.sigmoid(self.z1)
            print(self.a1)
            # calculate output for output layer
            self.z2 = np.dot(self.a1, self.W2)
            print(self.z2)
            self.a2 = self.sigmoid(self.z2)
            print(self.a2)
            pred.append(self.a2)
        # return y - pred
        return pred
        
```
